## 곡선 그리기

y = x**2 에 대한 그래프

x = 0 부터 99
y = x**2 인 관계



```python
import matplotlib.pyplot as plt

x = range(100)
y = [value **2 for value in x]

plt.plot(x,y)
plt.show()
```

```python
import math
T = range(100)

x = [x/100. * 10 for x in T]

y = [math.sin(value) for value in x]

plt.plot(x,y)
plt.show()
```

## 2파이 주기 동안 sin 삼각함수 그래프

np.linspace : 특정 두개의 수 사이에 균등하게 분리하는 1차원 배열

```python
import math
T = range(100)

x = [(2*math.pi *t)/len(T) for t in T]

y = [math.sin(value) for value in x]

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

##  2차원 함수에 대한 그래프 그리기

y = (x-1)**2

이 함수에 대한 그래프

```python
x = np.linspace(-3,2,200)
y = x**2 - 2*x + 1.

plt.plot(x,y)
plt.show()
```

## 여러 개 plot 그리기

plot 을 그리면 그리는 만큼 선을 표시

선은 파란색, 초록색 등으로 선을 그림

```python
x = np.linspace(0, 2*np.pi, 100) 
ya = np.sin(x)

yb = np.cos(x)

yc = np.sin(x) * -1
yd = np.cos(x) * -1
plt.plot(x,ya)
plt.plot(x,yb)
plt.plot(x,yc)
plt.plot(x,yd)
plt.yticks([-1,0,1])

plt.show()
```

```python
x = np.linspace(0, 2.*np.pi , 100) 

yc = np.tan(x)

plt.plot(x,yc)
plt.yticks([-2.,0,2.])

plt.show()
```

##  순서별로 그래프를 랜더링

첫번째 plot 함수를 그리고 두번째 함수에서 plot함수를 호출하므로 차례대로 그래프가 표시 됨

```python
gr = []
def plot_slope(x,y) :
    xs = x[1:] - x[:-1]
    ys = y[1:] - y[:-1]
    
    i =plt.plot(x[1:],ys/xs)
    gr.append(i)
    print(" func call ",i)
    
x = np.linspace(-3,3, 100)
y = np.exp(-x **2)

#first graph
i = plt.plot(x,y)
print(" plot ",i)
gr.append(i)

#second graph
plot_slope(x,y)

plt.show()
print(gr)
```
