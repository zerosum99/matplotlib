## 파일을 읽고 변환하기

1. 파일로 읽은 데이터는 전부 문자열로 처리
2. 숫자로 계산하려면 치환해야 함
3. 파일로 읽은 라인을 분리
    문자열.split() 메소드로 빈공칸을 구분으로 2개로 분리
    리스트 내포 [ float(s) for s in line.split() ] 로 리스트를 만듬
    
4. x,y 변수에 할당해서 도표를 그림

```python
import matplotlib.pyplot as plt

x, y = [], [] 
for line in open('my_data.txt','r') : 
    values = [ float(s) for s in line.split()] 
    print("values ", values)
    x.append(values[0])
    y.append(values[1])
    
plt.plot(x,y)
plt.show()
```

## 벡터화 계산(broadcast)


벡터와 스칼라, 벡터와 벡터간 계산을 위해서는 양쪽이 동일한 차수가 맞도록 확장되어 계산

example : 

x110 = np.array([[1,2,3],[4,5,6]])

print(x110 + x110.T)

## 다중 막대 그리프 그리기

데이터는 linspace으로 1차원 원소가 총 30개를 만들고 reshape(10,3)으로 2차원 배열을 만듬 

```python
data = np.linspace(1,10,30).reshape(10,3)

plt.bar(np.arange(10),data[:,0], width=0.25)
plt.bar(np.arange(10)+0.25,data[:,1], color='r',width=0.25)
plt.bar(np.arange(10)+0.50,data[:,2], color='g',width=0.25)
plt.show()
```

## 막대 그래프를 통합해서 차트 그리기

기존 막대 그래프를 bottom으로 처리해서 그리면 됨

```python
data = np.linspace(1,10,30).reshape(10,3)

plt.bar(np.arange(10),data[:,0])
plt.bar(np.arange(10),data[:,1], color='r', bottom=data[:,0])

plt.show()
```

##  막대그래프에 3개를 그릴때 주의사항

bottom에 첫번째와 두번째를 더한 위치부터 그려야 중첩되지 않는다.


```python
data = np.linspace(0,30,30).reshape(10,3)

plt.bar(np.arange(10),data[:,0])
plt.bar(np.arange(10),data[:,1], color='r', bottom=data[:,0])
plt.bar(np.arange(10),data[:,2], color='g', bottom= data[:,0]+ data[:,1])
plt.show()
```

## 수평막대를 2개 연결해서 막대차트 그리기

left에 첫번째 데이터를 넣으면 처음 막대 그래프에 추가해서 그림

```python
data = np.linspace(1,10,30).reshape(10,3)

plt.barh(np.arange(10),data[:,0])
plt.barh(np.arange(10),data[:,1], color='r', left=data[:,0])

plt.show()
```

```python
data = np.linspace(0,10,30).reshape(10,3)

plt.barh(np.arange(10),data[:,0])
plt.barh(np.arange(10),data[:,1], color='r', left=data[:,0])
plt.barh(np.arange(10),data[:,2], color='g', left=data[:,0]+data[:,1])

plt.show()
```

## 수평으로 양방향 막대 그래프 그리기

 첫번째는 정상 방향으로 그리고 두번째는 -1을 곱해서 반대방향으로 그리면 2개가 동시에 그려짐

# Plot layouts


subplot( 행, 열, 그래프사이즈) 를 나타내며 행과 열은 그래프를 표시함

## 1차원 2개의 그래프 그리기

```python
fig, axes = plt.subplots(2,figsize=(4,4))
axes[0].plot(x,y, 'r')
axes[1].plot(x,y, 'g')

for ax in axes.ravel():
    ax.margins(0.05)
plt.show()
```

## 2 행 2열로 그리기

```python

fig, axes = plt.subplots(2,2,figsize=(4,4))
axes[0,0].plot(x,y, 'r')
axes[0,1].plot(x,y, 'g')
axes[1,0].plot(x,y, 'b')
axes[1,1].plot(x,y, 'k')
for ax in axes.ravel():
    ax.margins(0.05)
plt.show()
```

## 3행 2열로 그리기 

```python

fig, axes = plt.subplots(3,2,figsize=(4,4))
axes[0,0].plot(x,y, 'r')
axes[0,1].plot(x,y, 'g')
axes[1,0].plot(x,y, 'b')
axes[1,1].plot(x,y, 'k')
axes[2,0].plot(x,y, 'c')
axes[2,1].plot(x,y, 's')
for ax in axes.ravel():
    ax.margins(0.05)
plt.show()
```

## 행과 열을 span 으로 조정하기 



```python
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0), colspan=2)
axes = [ax1, ax2, ax3, ax4]
colors = ['r', 'g', 'b', 'k']
for ax, c in zip(axes, colors):
    ax.plot(x, y, c)
    ax.margins(0.05)
plt.tight_layout()
plt.show()
```
