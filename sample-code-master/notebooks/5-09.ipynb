## 第5章 Matplotlibを使いこなそう

### 5-9: ファイル出力

```python
# リスト5.9.1：ファイルの出力
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
dat = [1, 3, 2]
ax.plot(dat)
fname = "save_fig.svg"
plt.savefig(fname)
plt.show()
```

```python
# リスト5.9.2：出力時にフィギュアオブジェクトの書式を設定
fig = plt.figure()
ax = fig.add_subplot(111)
dat = [1, 3, 2]
ax.plot(dat)
fname = "mpl_save_fig_prop_setting.svg"
plt.savefig(
    fname,
    dpi=300,
    facecolor="grey",
    edgecolor="k",
    linewidth=10,
    frameon=True,
    tihgt_layout=False,
)
plt.show()
```

```python
import numpy as np

# リスト5.9.3：サブプロットサイズの調整
x = np.arange(0, 15, 0.2)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure(figsize=(4, 3), edgecolor="b", linewidth=5, frameon=True)
ax = fig.add_subplot(111)
ax.plot(x, y, label="sin")
ax.plot(x, z, label="cos")
ax.set_xlabel("x value")
ax.set_ylabel("y value")
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
# サブプロットの位置の調整
fig.subplots_adjust(left=0.2, bottom=0.2, right=0.75, top=0.9)
fname = "save_fig_adjust.svg"
plt.savefig(fname, dpi=300, edgecolor="b", linewidth=2, frameon=True)
plt.show()
```
